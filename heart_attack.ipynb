```python
# Add Column age_group
bins = [ 30, 40, 50, 60, 70, 80]
labels = ['30-39', '40-49','50-59','60-69','70+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)
```

### sysBP --> Systolic Blood Pressure
### diaBP --> Diastolic Blood Pressure

```python
# Categorize Data into Systolic Blood Pressure Levels
bins = [0, 120, 130, 140, 160, 180, 300]
labels = ["Normal", "Elevated", "High-Normal", "Hypertension Stage 1", "Hypertension Stage 2", "Hypertensive Crisis"]
df['Sys_Cat'] = pd.cut(df['sysBP'], bins=bins, labels=labels, right=False)
```

```python
# Categorize Data into Diastolic Blood Pressure Levels 
bins = [0, 80, 85, 90, 100, 110, 180]
labels = ["Normal", "Elevated" ,"High-Normal", "Hypertension Stage 1", "Hypertension Stage 2", "Hypertensive Crisis"]
df['Dia_Cat'] = pd.cut(df['diaBP'], bins=bins, labels=labels, right=False)
```

```python
df['Sys_Cat'].value_counts()
```

```python
df['Dia_Cat'].value_counts()
```

```python
df.head()
```

```python
risk_order = ["Normal", "Elevated", "High-Normal", "Hypertension Stage 1", "Hypertension Stage 2", "Hypertensive Crisis"]
def max_risk(row):
    sys_idx = risk_order.index(row['Sys_Cat'])
    dia_idx = risk_order.index(row['Dia_Cat'])
    return risk_order[max(sys_idx, dia_idx)]
df['BP_Cat'] = df.apply(max_risk, axis=1)
```

```python
df.head()
```

```python
# Categorize Data into Heart Rate Levels 

# Bradycardia(Slow) ---- Tachycardia(Fast)  
bins = [0, 60, 101, 150]
labels = ['Bradycardia', 'Normal', 'Tachycardia'] 
df['HeartRate_Cat'] = pd.cut(df['heartRate'], bins=bins, labels=labels, right=False)
```

```python
df['HeartRate_Cat'].value_counts()
```

```python
# Categorize Data into Glucose Levels 
bins = [0, 70, 100, 126, 400]
labels = ['Hypoglycemia', 'Normal', 'Prediabetes', 'Diabetes'] 
df['Glucose_Cat'] = pd.cut(df['glucose'], bins=bins, labels=labels, right=False)
```

### Detect Outliers (UpNoraml Values)

```python
# outliers by IQR for column 'glucose'


# Calculate IQR
q1 = df['glucose'].quantile(0.25)
q3 = df['glucose'].quantile(0.75)
iqr = q3 - q1

# Define bounds
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr


outliers = df[(df['glucose'] < lower_bound) | (df['glucose'] > upper_bound)]
print(f"New Number of raws in 'glucose': {len(outliers)}")

```

### glucose (Blood Sugar): This column has the most outliers (262 cases).

***Observation*** : The maximum value is 394 mg/dL, while the upper bound for "normal" data distribution in this set is around 111 mg/dL.

***Insight*** : These high values likely represent patients with uncontrolled diabetes. Since this is a heart disease study, these are scientifically important values, not necessarily errors.

```python
# outliers by IQR for systolic_BP (sysBP)

# Calculate IQR
q1 = df['sysBP'].quantile(0.25)
q3 = df['sysBP'].quantile(0.75)
iqr = q3 - q1

# Define bounds
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = df[(df['sysBP'] < lower_bound) | (df['sysBP'] > upper_bound)]
print(f"New Number of raws in 'sysBP': {len(outliers)}")
```
